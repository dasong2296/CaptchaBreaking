```python
import os
import cv2 as cv
from pytesseract import image_to_string
from PIL import Image
from PIL import ImageFilter
from PIL import ImageOps
from PIL import ImageEnhance
import timeit
import numpy as np
import pandas as pd
from pathlib import Path
import numpy as np
from matplotlib import pyplot as plt

data_dir = Path("./OCR_DEMO_DATASET/")
images = sorted(list(map(str, list(data_dir.glob("*.jpg")))))
labels = [img.split(os.path.sep)[-1].split(".jpg")[0] for img in images]
characters = set(char for label in labels for char in label)

print("Number of images found: ", len(images))
print("Number of labels found: ", len(labels))
print("Number of unique characters: ", len(characters))
print("Characters present: ", characters)

def recognize_text_2(i):
    images = []
    img_ori = cv.imread(i)
    height, width, _ = img_ori.shape
    #  edge preserving filter denoising 
    image = cv.cvtColor(img_ori, cv.COLOR_BGR2GRAY)
    image = cv.medianBlur(image,5)
    
    image = cv.equalizeHist(image)

    ret, binary = cv.threshold(image, 127, 255, cv.THRESH_BINARY)
    th2 = cv.adaptiveThreshold(image,255,cv.ADAPTIVE_THRESH_MEAN_C,\
            cv.THRESH_BINARY,11,2)
    th3 = cv.adaptiveThreshold(image,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)
    print(f' binarization adaptive threshold ï¼š{ret}')
    images= [img_ori, binary, th2, th3]
    titles = ['Original Image', 'Global Thr (v=127)',
            'Adaptive Mean Thr', 'Adaptive Gaussian Thr']
    
    vertical_hist_th2 = th2.shape[0] - np.sum(th2,axis=0,keepdims=True)/255
    vertical_hist_binary = binary.shape[0] - np.sum(binary,axis=0,keepdims=True)/255
    vertical_hist_th3 = th3.shape[0] - np.sum(th3,axis=0,keepdims=True)/255
    
    blankImage_binary = np.zeros_like(image)
    blankImage_th2 = np.zeros_like(image)
    blankImage_th3 = np.zeros_like(image)
    
    for i, value in enumerate(vertical_hist_th2[0]):
        cv.line(blankImage_th2, (i, 0), (i, height-int(value)), (255, 255, 255), 1)
    
    for i, value in enumerate(vertical_hist_binary[0]):
        cv.line(blankImage_binary, (i, 0), (i, height-int(value)), (255, 255, 255), 1)
    
    for i, value in enumerate(vertical_hist_th3[0]):
        cv.line(blankImage_th3, (i, 0), (i, height-int(value)), (255, 255, 255), 1)
    
    for i in range(4):
        plt.subplot(4,2,i+1), plt.imshow(images[i],'gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])

    plt.subplot(4,2, 6), plt.imshow(blankImage_binary)
    plt.title("Binary Vertical Histogram")
    plt.subplot(4,2, 7), plt.imshow(blankImage_th2)
    plt.title("Mean Thr Vertical Histogram")
    plt.subplot(4,2, 8), plt.imshow(blankImage_th3)
    plt.title("Gaussian Thr Vertical Histogram")
    
    
    plt.show()
    
count = 0
for i in images:
    if (count > 4):
        break
    count += 1
    pred = recognize_text_2(i)
    print(pred, (i.split('/')[-1]).split('.')[0])
```
